# NumPy Arrays — AI Engineering Interview Prep

Covers: array creation, indexing, broadcasting, linear algebra, performance.

```python
import numpy as np
import time

np.random.seed(42)
print("NumPy:", np.__version__)
```

## 1. Array Creation

```python
# Different creation methods
a = np.array([1, 2, 3, 4, 5])
b = np.zeros((3, 4))
c = np.ones((2, 3), dtype=np.float32)
d = np.eye(4)
e = np.arange(0, 20, 2)
f = np.linspace(0, 1, 11)
g = np.random.randn(3, 3)

print(f"1D array:   {a}")
print(f"zeros:      shape={b.shape}, dtype={b.dtype}")
print(f"ones:       dtype={c.dtype}")
print(f"arange:     {e}")
print(f"linspace:   {f}")
print(f"random:     \n{g.round(3)}")
```

## 2. Indexing & Slicing

```python
A = np.arange(24).reshape(4, 6)
print("Array A:")
print(A)

print(f"\nA[1, 3]    = {A[1, 3]}")
print(f"A[0:2, 2:5] =\n{A[0:2, 2:5]}")
print(f"A[:, -1]   = {A[:, -1]}     # last column")
print(f"A[::2, :]  =\n{A[::2, :]}   # every other row")

# Boolean indexing
mask = A > 12
print(f"\nA[A>12]: {A[mask]}")

# Fancy indexing
rows = [0, 2, 3]
print(f"Rows 0,2,3:\n{A[rows, :]}")
```

## 3. Array Operations & Broadcasting

```python
# Broadcasting rules: shapes are compatible if they're equal or one of them is 1
a = np.array([[1], [2], [3]])   # shape (3,1)
b = np.array([10, 20, 30, 40])  # shape (4,)
result = a + b  # broadcasts to (3,4)
print(f"a shape: {a.shape}, b shape: {b.shape}")
print(f"a + b shape: {result.shape}")
print(result)

# Normalize each row (subtract mean, divide by std)
X = np.random.randn(100, 5)
X_norm = (X - X.mean(axis=0)) / X.std(axis=0)
print(f"\nRow-normalized: mean≈{X_norm.mean(axis=0).round(10)}, std={X_norm.std(axis=0).round(4)}")
```

```python
# Vectorized operations vs Python loop
n = 1_000_000
a = np.random.randn(n)
b = np.random.randn(n)

# NumPy dot product
t0 = time.perf_counter()
result_np = np.dot(a, b)
t_np = time.perf_counter() - t0

# Python loop
t0 = time.perf_counter()
result_py = sum(x*y for x,y in zip(a, b))
t_py = time.perf_counter() - t0

print(f"NumPy:  {result_np:.4f}  time={t_np*1000:.2f}ms")
print(f"Python: {result_py:.4f}  time={t_py*1000:.2f}ms")
print(f"Speedup: {t_py/t_np:.0f}x")
```

## 4. Aggregations & Axis Operations

```python
X = np.random.randint(0, 10, (4, 5))
print("X:")
print(X)
print(f"\nsum():        {X.sum()}      # all elements")
print(f"sum(axis=0):  {X.sum(axis=0)}  # column sums")
print(f"sum(axis=1):  {X.sum(axis=1)}  # row sums")
print(f"mean(axis=0): {X.mean(axis=0).round(2)}")
print(f"argmax:       row={X.argmax(axis=1)}  # index of max in each row")
print(f"cumsum:       {X[0].cumsum()}")
```

## 5. Linear Algebra Operations

```python
A = np.array([[2., 1., 0.],
              [1., 3., 1.],
              [0., 1., 2.]])

# Matrix operations
print("Matrix A:")
print(A)
print(f"\nA @ A.T:\n{(A @ A.T).round(3)}")
print(f"\nTrace: {np.trace(A)}")
print(f"Det:   {np.linalg.det(A):.4f}")
print(f"Rank:  {np.linalg.matrix_rank(A)}")
print(f"Norm:  {np.linalg.norm(A):.4f}  (Frobenius)")

# Solve Ax = b
b = np.array([5., 10., 3.])
x = np.linalg.solve(A, b)
print(f"\nSolution x: {x}")
print(f"Verify Ax = {A @ x}  ≈ {b}")
```

## 6. Useful Tricks

```python
# reshape, ravel, flatten
a = np.arange(12)
print("reshape:", a.reshape(3, 4))
print("reshape -1:", np.arange(24).reshape(2, -1).shape)  # -1 infers

# Stack / concatenate
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print("vstack:", np.vstack([a, b]))
print("hstack:", np.hstack([a, b]))

# where — conditional element selection
x = np.array([-2, -1, 0, 1, 2])
result = np.where(x >= 0, x, 0)   # ReLU!
print("ReLU:", result)

# unique with counts
arr = np.array([1, 2, 2, 3, 1, 4, 3, 3])
vals, counts = np.unique(arr, return_counts=True)
print(dict(zip(vals, counts)))
```

## Interview Tips

- **Broadcasting**: shapes align from the right; size-1 dims expand. Master this to avoid loops.
- **Views vs copies**: slicing returns a view; boolean/fancy indexing returns a copy. Use `.copy()` explicitly.
- **dtype matters**: int32 vs float64 affects memory 2× and computation speed.
- **axis=0** operates along rows (produces result per column); **axis=1** operates along columns (result per row).
- `np.where(cond, a, b)` is vectorised ternary — essential for ML feature transforms.
- Always prefer NumPy vectorized ops over Python loops — typically 10-100× faster.
