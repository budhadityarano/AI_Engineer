## 1. Variables & Data Types

```python
# Core Python types
integer = 42
flt = 3.14159
string = "hello"
boolean = True
nothing = None
cmplx = 2 + 3j
byt = b"bytes"

for name, val in [("int", integer), ("float", flt), ("str", string),
                   ("bool", boolean), ("None", nothing), ("complex", cmplx)]:
    print(f"{name:8s}: {repr(val):20s} | type={type(val).__name__}")
```

```python
# isinstance is preferred over type() for subclass-aware checks
print(isinstance(True, int))   # True — bool is subclass of int!
print(isinstance(3.14, (int, float)))

# Type casting
print(int("42"), float("3.14"), str(99), bool(0), bool(""), bool("x"))
print(list("abc"), tuple([1,2,3]), set([1,1,2,3]))
```

## 2. String Operations

```python
s = "  Hello, World!  "

# Common methods
print(s.strip())                      # remove whitespace
print(s.strip().lower())
print(s.strip().replace(",", ""))
print(",".join(["a", "b", "c"]))
print("a,b,c".split(","))
print("hello".count("l"))
print("hello".find("ll"))             # index or -1
print("hello".startswith("he"))
```

```python
# String formatting
name, score = "Alice", 98.5
print(f"{name} scored {score:.1f}")                  # f-string (preferred)
print("{} scored {:.1f}".format(name, score))        # .format()
print("%s scored %.1f" % (name, score))              # %-style (legacy)

# Slicing: s[start:stop:step]
s = "abcdefgh"
print(s[2:5], s[::-1], s[::2], s[-3:])
```

## 3. Control Flow

```python
# if / elif / else + ternary
score = 85
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "F"
print(f"Grade: {grade}")

# Ternary (conditional expression)
label = "pass" if score >= 60 else "fail"
print(label)
```

```python
# for loops
fruits = ["apple", "banana", "cherry"]
for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")

# zip
names = ["Alice", "Bob"]
scores = [95, 87]
for n, s in zip(names, scores):
    print(f"{n}: {s}")

# loop else (runs if loop didn't break)
for x in range(5):
    if x == 10:
        break
else:
    print("No break occurred")
```

```python
# while, break, continue
n = 10
total = 0
while n > 0:
    if n % 3 == 0:
        n -= 1
        continue    # skip multiples of 3
    total += n
    n -= 1
print(f"Sum (skip mult of 3): {total}")
```

## 4. Functions

```python
# *args, **kwargs, defaults, multiple return values
def stats(label, *numbers, round_to=2):
    """Returns (mean, min, max) for the given numbers."""
    mean = sum(numbers) / len(numbers)
    return round(mean, round_to), min(numbers), max(numbers)

mean, lo, hi = stats("test", 4, 7, 2, 9, 1, round_to=1)
print(f"mean={mean}, min={lo}, max={hi}")

# **kwargs
def build_profile(**kwargs):
    return {k: v for k, v in kwargs.items()}

print(build_profile(name="Alice", role="engineer", yoe=5))
```

```python
# Closures and nested functions
def make_multiplier(factor):
    def multiply(x):
        return x * factor   # captures 'factor' from outer scope
    return multiply

double = make_multiplier(2)
triple = make_multiplier(3)
print(double(5), triple(5))
```

## 5. Comprehensions

```python
# List comprehensions
squares = [x**2 for x in range(10)]
evens = [x for x in range(20) if x % 2 == 0]
flat = [val for row in [[1,2],[3,4],[5,6]] for val in row]  # nested
print(squares)
print(evens)
print(flat)

# Dict comprehension
word_len = {word: len(word) for word in ["cat", "elephant", "ox"]}
print(word_len)

# Set comprehension
unique_mods = {x % 5 for x in range(20)}
print(unique_mods)
```

## 6. Functional Tools

```python
nums = [1, -3, 4, -2, 7, -1, 0, 5]

# lambda
square = lambda x: x ** 2

# map, filter
squared = list(map(square, nums))
positives = list(filter(lambda x: x > 0, nums))
print("squared:", squared)
print("positives:", positives)

# sorted with key
words = ["banana", "fig", "apple", "date"]
print(sorted(words, key=len))           # by length
print(sorted(words, key=lambda w: w[-1]))  # by last char

# any, all
print(any(x > 0 for x in nums))
print(all(x > 0 for x in nums))
```

## 7. Exception Handling

```python
# Custom exception
class ValidationError(ValueError):
    pass

def parse_age(s):
    try:
        age = int(s)
    except ValueError:
        raise ValidationError(f"Cannot parse age from: {s!r}")
    if age < 0 or age > 150:
        raise ValidationError(f"Age {age} out of range")
    return age

for val in ["25", "abc", "-5", "200"]:
    try:
        print(f"parse_age({val!r}) = {parse_age(val)}")
    except ValidationError as e:
        print(f"ValidationError: {e}")
```

```python
# try / except / else / finally
def safe_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        return None
    except TypeError as e:
        raise ValueError(f"Invalid types: {e}")
    else:
        return result   # runs only if no exception
    finally:
        pass            # always runs (cleanup)

print(safe_divide(10, 2))
print(safe_divide(10, 0))
```

## 8. Useful Built-ins

```python
nums = [3, 1, 4, 1, 5, 9, 2, 6]

print("len:", len(nums))
print("sum:", sum(nums))
print("min/max:", min(nums), max(nums))
print("sorted:", sorted(nums))
print("reversed:", list(reversed(nums)))
print("abs(-7):", abs(-7))
print("round(3.567, 2):", round(3.567, 2))
print("divmod(17, 5):", divmod(17, 5))     # (quotient, remainder)
print("pow(2, 10):", pow(2, 10))

# zip_longest — pad shorter iterables
from itertools import zip_longest
a = [1, 2, 3]
b = ["a", "b"]
print(list(zip_longest(a, b, fillvalue=0)))
```

## Interview Tips

- `is` checks identity; `==` checks equality. `None` comparisons: always use `is None`.
- Mutable defaults in functions are a common bug: use `def f(lst=None): lst = lst or []`.
- List comprehensions are faster than equivalent `for`+`append` loops.
- `enumerate(iterable, start=1)` gives 1-indexed counters.
- `dict.get(key, default)` avoids KeyError without try/except.
- Know the difference: `list.sort()` in-place; `sorted()` returns new list.
