# Arrays & Strings — Coding Interview Prep

Two Sum, sliding window, prefix sums, and other core patterns.

## 1. Two Sum — Hash Map O(n)

```python
def two_sum(nums, target):
    seen = {}  # value → index
    for i, n in enumerate(nums):
        complement = target - n
        if complement in seen:
            return [seen[complement], i]
        seen[n] = i
    return []

assert two_sum([2, 7, 11, 15], 9) == [0, 1]
assert two_sum([3, 2, 4], 6) == [1, 2]
assert two_sum([3, 3], 6) == [0, 1]
print("two_sum: all tests passed")
```

## 2. Three Sum — Sort + Two Pointers O(n²)

```python
def three_sum(nums):
    nums.sort()
    result = []
    for i in range(len(nums) - 2):
        if i > 0 and nums[i] == nums[i-1]:  # skip duplicates
            continue
        lo, hi = i + 1, len(nums) - 1
        while lo < hi:
            s = nums[i] + nums[lo] + nums[hi]
            if s == 0:
                result.append([nums[i], nums[lo], nums[hi]])
                while lo < hi and nums[lo] == nums[lo+1]: lo += 1
                while lo < hi and nums[hi] == nums[hi-1]: hi -= 1
                lo += 1; hi -= 1
            elif s < 0: lo += 1
            else: hi -= 1
    return result

assert sorted(three_sum([-1, 0, 1, 2, -1, -4])) == [[-1, -1, 2], [-1, 0, 1]]
assert three_sum([0, 0, 0]) == [[0, 0, 0]]
print("three_sum: all tests passed")
```

## 3. Sliding Window — Max Sum Subarray of size k

```python
def max_subarray_sum_k(nums, k):
    window = sum(nums[:k])
    max_sum = window
    for i in range(k, len(nums)):
        window += nums[i] - nums[i - k]
        max_sum = max(max_sum, window)
    return max_sum

assert max_subarray_sum_k([2, 1, 5, 1, 3, 2], 3) == 9
assert max_subarray_sum_k([2, 3, 4, 1, 5], 2) == 7

# Variable window: longest substring without repeating characters
def length_of_longest_substring(s):
    char_idx = {}
    max_len = left = 0
    for right, c in enumerate(s):
        if c in char_idx and char_idx[c] >= left:
            left = char_idx[c] + 1
        char_idx[c] = right
        max_len = max(max_len, right - left + 1)
    return max_len

assert length_of_longest_substring("abcabcbb") == 3
assert length_of_longest_substring("bbbbb") == 1
assert length_of_longest_substring("pwwkew") == 3
print("Sliding window: all tests passed")
```

## 4. Prefix Sum — Range Sum Queries O(1)

```python
class NumArray:
    def __init__(self, nums):
        self.prefix = [0] * (len(nums) + 1)
        for i, n in enumerate(nums):
            self.prefix[i+1] = self.prefix[i] + n

    def sum_range(self, left, right):
        return self.prefix[right+1] - self.prefix[left]

na = NumArray([-2, 0, 3, -5, 2, -1])
assert na.sum_range(0, 2) == 1
assert na.sum_range(2, 5) == -1
assert na.sum_range(0, 5) == -3

# Subarray sum equals k (count subarrays)
def subarray_sum_k(nums, k):
    count = 0
    prefix_sum = 0
    counts = {0: 1}  # prefix_sum → frequency
    for n in nums:
        prefix_sum += n
        count += counts.get(prefix_sum - k, 0)
        counts[prefix_sum] = counts.get(prefix_sum, 0) + 1
    return count

assert subarray_sum_k([1, 1, 1], 2) == 2
assert subarray_sum_k([1, 2, 3], 3) == 2
print("Prefix sum: all tests passed")
```

## 5. Kadane's Algorithm — Maximum Subarray O(n)

```python
def max_subarray(nums):
    max_ending = max_global = nums[0]
    for n in nums[1:]:
        max_ending = max(n, max_ending + n)
        max_global = max(max_global, max_ending)
    return max_global

assert max_subarray([-2, 1, -3, 4, -1, 2, 1, -5, 4]) == 6
assert max_subarray([1]) == 1
assert max_subarray([5, 4, -1, 7, 8]) == 23
print("Kadane: all tests passed")
```

## 6. String Patterns

```python
from collections import Counter

# Valid anagram
def is_anagram(s, t):
    return Counter(s) == Counter(t)

assert is_anagram("anagram", "nagaram")
assert not is_anagram("rat", "car")

# Group anagrams
def group_anagrams(strs):
    groups = {}
    for s in strs:
        key = tuple(sorted(s))
        groups.setdefault(key, []).append(s)
    return list(groups.values())

result = group_anagrams(["eat", "tea", "tan", "ate", "nat", "bat"])
assert sorted(sorted(g) for g in result) == [['ate', 'eat', 'tea'], ['bat'], ['nat', 'tan']]

# Minimum window substring
def min_window(s, t):
    if not t or not s: return ""
    need = Counter(t)
    formed = 0
    window = {}
    left = 0
    min_len = float('inf')
    min_start = 0
    for right, c in enumerate(s):
        window[c] = window.get(c, 0) + 1
        if c in need and window[c] == need[c]:
            formed += 1
        while formed == len(need):
            if right - left + 1 < min_len:
                min_len = right - left + 1
                min_start = left
            lc = s[left]
            window[lc] -= 1
            if lc in need and window[lc] < need[lc]:
                formed -= 1
            left += 1
    return s[min_start:min_start+min_len] if min_len != float('inf') else ""

assert min_window("ADOBECODEBANC", "ABC") == "BANC"
assert min_window("a", "a") == "a"
print("String patterns: all tests passed")
```

## Patterns Summary

| Pattern | Time | Space | Use When |
|---------|------|-------|----------|
| Hash Map | O(n) | O(n) | Two Sum, frequency counts |
| Two Pointers | O(n) | O(1) | Sorted array, palindrome |
| Sliding Window | O(n) | O(k) | Subarray/substring conditions |
| Prefix Sum | O(n) precompute, O(1) query | O(n) | Range sum queries |
| Kadane's | O(n) | O(1) | Max subarray sum |
